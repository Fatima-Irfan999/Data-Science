```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/canada_per_capita_income.csv")
```

```python
df
```

```python
plt.scatter(df['year'], df['per capita income (US$)'])
plt.xlabel("Year")
plt.ylabel("Income ($)")
plt.title('Scatter Plot of Per Capita Income over Time')
```

```python
x = df['year']
y = df['per capita income (US$)']
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x.values.reshape(-1, 1), y)
```

```python
slope = model.coef_[0]
intercept = model.intercept_
```

```python
print(slope)
print(intercept)
```

```python
regression_line = [slope * xi + intercept for xi in x]
```

```python
plt.scatter(x, y, label='Data Points')
plt.plot(x, regression_line, color='red', label='Regression Line')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Scatter Plot with Linear Regression')
plt.legend()
plt.show()
```

```python
user_year = int(input("Enter a year to predict the per capita income: "))
predicted_income = model.predict([[user_year]])
print(f"The predicted income for the year {user_year} is ${predicted_income[0]:.2f}")
```
